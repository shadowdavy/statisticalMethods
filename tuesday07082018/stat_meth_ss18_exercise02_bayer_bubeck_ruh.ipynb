### 2.1 Farthers and sons on the career ladder (analytic problem) {-}
Suppose that of the father’s generation 10 % are in U , 40 %
in M , and 50 % in L, i.e., $P(U_1)=10\%$, $P(M_1)=40\%$ \& $P(L_1)=50\%$.

#### a: What is the probability that a son in the next generation is in U? Analoguously in M , or L?

Probability that a son in the next generation is in U given by law of total probability because all the possible events {$U_1$,$M_1$,$L_1$} are mutually exclusive:
$$P(U_2)=P(U_2|U_1)P(U_1)+P(U_2|M_1)P(M_1)+P(U_2|L_1)P(L_1)=45\%\cdot10\%+5\%\cdot40\%+1\%\cdot50\%=7\%$$

Probability that a son in the next generation is in M (analogous to U):
$$P(M_2)=P(M_2|U_1)P(U_1)+P(M_2|M_1)P(M_1)+P(M_2|L_1)P(L_1)=48\%\cdot10\%+70\%\cdot40\%+50\%\cdot50\%=57.8\%$$

Probability that a son in the next generation is in L (analogous to U):
$$P(L_2)=P(L_2|U_1)P(U_1)+P(L_2|M_1)P(M_1)+P(L_2|L_1)P(L_1)=7\%\cdot10\%+25\%\cdot40\%+49\%\cdot50\%=35.2\%$$

#### b: If a son has occupational status $U_2$, what is the probability that his father had occupational status $U_1$? Analoguously $M_1$, or $L_1$?

Use Bayes' theorem: $$P(U_1|U_2)=\frac{P(U_2|U_1)P(U_1)}{P(U_2)}=\frac{45\%\cdot10\%}{7\%}\simeq64.3\%$$

Analogous to $U_1$: $$P(M_1|U_2)=\frac{P(U_2|M_1)P(M_1)}{P(U_2)}=\frac{5\%\cdot40\%}{7\%}\simeq28.6\%$$

Analogous to $U_1$: $$P(L_1|U_2)=\frac{P(U_2|L_1)P(L_1)}{P(U_2)}=\frac{1\%\cdot50\%}{7\%}\simeq7.1\%$$

### 2.2 The unavoidable drawing from a box (or urn) (analytic problem) {-}

#### a: What is the probability that you draw a black marble on the first draw?

$$P(B_1)=\frac{M}{N}$$

#### b: What is the probability that you draw a black marble on the second draw if you know that a black marble was drawn on the first draw and not replaced?

After we removed one black marble from the box on the first draw we are left with $M-1$ black marbles and $N-1$ marbles in total on the second draw, therefore, the probability to draw a black marble on the second draw if we know that a black marble was drawn on the first draw and that this first marble is not replaced is given by
$$P(B_2|B_1)=\frac{M-1}{N-1}$$

#### c: What is now the probability that you draw a black marble on the second draw if you are not told what colour was picked on the first draw (and that first marble was also not replaced)? Hint: The result is counter-intuitive. Apply the law of total probability!

Probability to draw a white marble on the first draw: $$P(W_1)=\frac{N-M}{N}$$

After one white marble is removed on the first draw $N-m-1$ white marbles and $N-1$ marbles in total are left, hence the probability to draw a black marble on the second draw if we know that a white marble was drawn on the first draw and that this first marble is not replaced is: $$P(B_2|W_1)=\frac{M}{N-1}$$

Application of the law of total probability yields the required probability to draw a black marble on the second draw if you we are not told what colour was picked on the first draw (and that first marble was also not replaced): $$P(B_2)=P(B_2|B_1)P(B_1)+P(B_2|W_1)P(W_1)=\frac{M-1}{N-1}\frac{M}{N}+\frac{M}{N-1}\frac{N-M}{N}=\frac{1}{N-1}\frac{M}{N}(M-1+N-M)$$
$$=\frac{M}{N}\frac{N-1}{N-1}=\frac{M}{N}$$

### 2.3 A small pictoral atlas of propability distributions  {-}

#### a: For each distribution write down its functional form

##### Poisson distribution

$$
P(k) = e^{-\lambda}\frac{\lambda^k}{k!} \mathrm{,}
$$

where $k$ is the number of times an event occurs and $k \geq 0$, and the average number of events is denoted by $\lambda$.

##### Beta distribution

$$
f(x;\alpha,\beta) =  \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}\, x^{\alpha-1}(1-x)^{\beta-1} \mathrm{,}
$$

where the $\Gamma$ function is given by

$$
\Gamma(z) = \int_0^\infty x^{z-1} e^{-x} \mathrm{,}
$$

and $\alpha,\beta > 0$ determine the shape of the Beta distribution.



##### $\chi^2$ distribution

$$
f(x;k) = \frac{x^{\frac{k}{2}-1}e^{-\frac{x}{2}}}{2^{\frac{k}{2}}\Gamma\left(\frac{k}{2}\right)}
$$

Here, $k$ is a positive integer and denotes the degrees of freedom.

#### b: Create a figure illustrating their dependence on parameters

```r
# Poisson distribution
# set sequence for different values of k and lambda, lambda in step of 2
stepLambda <- seq(1, 10, by=2)
k <- seq(0, 20)
# set colors
color <- c("blue", "brown", "red", "gold", "darkgreen", "darkmagenta", "aquamarine4",
           "chocolate3", "chartreuse3")
i = 1 # variable for color change
# save plot as png file
#png("poisson_distribution.png")
# plot poisson distribution for different values of k and lambda
plot(k, dpois(k, 1), col=color[1], main="") # lambda = 1
title(main = "Poisson distribution")
for (lambda in stepLambda) {
        points(k, dpois(k, lambda), col=color[i])
        lines(k, dpois(k, lambda), col=color[i])
        i = i + 1
}
# insert a legend
legend("topright", legend=stepLambda, inset=.05, title="lambda", lwd=2,
       lty=c(1, 1, 1, 1, 2), col=color)


# Beta distribution
# set sequence for different values of alpha and beta, (alpha, beta > 0)
alpha <- c(0.5, 1, 2, 5)
beta <- c(0.5, 1, 2, 3, 5)
# set colors
color <- c("blue", "brown", "red", "gold", "darkgreen", "darkmagenta", "aquamarine4",
           "chocolate3", "chartreuse3")
# save plot as png file
#png("beta_distribution.png")
# plot beta distribution for different values of alpha and beta
# plot not all variations of alpha and beta, take a few where alpha = beta,
# alpha > beta and alpha < beta to show dependency
x <- seq(0, 1, length=50)
# alpha = beta
plot(x, dbeta(x, alpha[1], beta[1]), col=color[1], ylim=c(0, 3), main="")
title(main = "Beta distribution")
lines(x, dbeta(x, alpha[1], beta[1]), col=color[1])
points(x, dbeta(x, alpha[3], beta[3]), col=color[2])
lines(x, dbeta(x, alpha[3], beta[3]), col=color[2])
# alpha > beta
points(x, dbeta(x, alpha[4], beta[2]), col=color[3])
lines(x, dbeta(x, alpha[4], beta[2]), col=color[3])
# alpha < beta
points(x, dbeta(x, alpha[2], beta[4]), col=color[4])
lines(x, dbeta(x, alpha[2], beta[4]), col=color[4])
points(x, dbeta(x, alpha[3], beta[5]), col=color[5])
lines(x, dbeta(x, alpha[3], beta[5]), col=color[5])
# insert a legend
legendValue = c("alpha=0.5, beta=0.5", "alpha=2, beta=2", "alpha=5, beta=1",
                "alpha=1, beta=3", "alpha=2, beta=5")
legend("topright", legend=legendValue, inset=.05, lwd=2,
       lty=c(1, 1, 1, 1, 2), col=color)


# Chi-squared distribution
# set sequence for different values of k (degrees of freedom)
k <- c(1, 2, 4, 6, 9)
# set colors
color <- c("blue", "brown", "red", "gold", "darkgreen", "darkmagenta", "aquamarine4",
           "chocolate3", "chartreuse3")
col = 1 # variable for color change
# save plot as png file
#png("chiSquared_distribution.png")
x <- seq(0, 20, length=50)
# plot chi squared distribution for different values of k
plot(x, dchisq(x, 1), col=color[1], ylim=c(0, 0.5), main="") # k = 1
title(main = "chi2 distribution")
for (i in k) {
    points(x, dchisq(x, i), col=color[col])
    lines(x, dchisq(x, i), col=color[col])
    col = col + 1
}
# insert a legend
legend("topright", legend=k, inset=.05, title="k", lwd=2,
       lty=c(1, 1, 1, 1, 2), col=color)
```

#### c: Write a brief note whether it is a continuous or discrete distribution, its domain of definition, expectation value and variance, and where it might play a role

##### Poisson distribution
* discrete distribution
* domain of definition: $k\geq0$
* expectation value = $\lambda$
* variance = $\lambda$
* It plays a role where a given number of events occur in a fixed interval of time or space if these events occur with a known constant rate and independently of the time since the last event. An example would be incoming photons observed on a detector and the appropriate probability to detect and not detect a photon.

##### Beta distribution
* family of continuous distributions
* domain of definition: [0,1] & parametrised by shape parameters $\alpha>0$, $\beta>0$
* expectation value = $\frac{\alpha}{\alpha+\beta}$
* variance = $\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$
* The beta distribution could play a role as an appropriate density function for parameters constrained to lie in the range 0–1.

##### $\chi^2$ distribution
* continuous distribution
* domain of definition: $x>0$ & $k\geq1$
* expectation value = $k$
* variance = $2k$
* Application: construction of confidence intervals & hypothesis testing
